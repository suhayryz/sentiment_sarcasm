```python
import pandas as pd
import nltk
from nltk.tag import pos_tag
import numpy as np
import csv
import re
```

```python
df = pd.read_csv("final/english1970_data(baseline).csv", header= 0, encoding="utf-8")
df.shape
```

```python
df.columns = ['comment','sarc_majority']
df['comment'] = df['comment'].str.lower().str.split()
```

```python
pd.set_option('display.max_colwidth', 1000)
df.head(10)
```

```python
pos_tagged_tokens = [nltk.pos_tag(t,tagset='universal') for t in df.comment]
pos_tagged_tokens
```

```python
df.comment= pos_tagged_tokens
df.head(10)
```

```python
df.comment = df.comment.apply(lambda x: [(t[0],) for t in x if t[1]=='ADJ' or t[1]=='ADV' or t[1]=='NOUN' or t[1]=='PRON' or t[1]=='VERB'])
df.ix[df.comment.apply(len) == 0, 'comment'] = [[np.nan]]
df
```

```python
df.to_csv('save4weka/eng1970_data(pos).csv', index=False)
```
