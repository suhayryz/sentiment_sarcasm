```python
# libraries
import pandas as pd
import nltk
from nltk.tag import pos_tag
import numpy as np
import csv
import re
```

```python
# load data
df = pd.read_csv('../datasets/english1970_data.csv', header= 0, encoding="utf-8")
df.shape
```

```python
# lowercase
df.columns = ['comment','sarc_majority']
df['comment'] = df['comment'].str.lower().str.split()
```

```python
# set max each column
pd.set_option('display.max_colwidth', 1000)
df.head(10)
```

```python
# tag with tagset
pos_tagged_tokens = [nltk.pos_tag(t,tagset='universal') for t in df.comment]
pos_tagged_tokens
```

```python
df.comment= pos_tagged_tokens
df.head(10)
```

```python
# only tagged taken
df.comment = df.comment.apply(lambda x: [(t[0],) for t in x if t[0]=='ADJ' or t[0]=='ADV' or t[0]=='NOUN' or t[0]=='PRON' or t[1]=='VERB'])
df.loc[df.comment.apply(len) == 0, 'comment'] = [[np.nan]]
df
```

```python
# save to 
df.to_csv('../datasets/english1970_data_pos.csv', index=False)
```
