### 7. Combine all feature dataset

```python
# libraries
from functools import reduce
import nltk
import pandas as pd
import numpy as np
import re
```

```python
# set max each column
pd.set_option('display.max_colwidth', 100)
```

```python
# load all data
df_syn = pd.read_csv('../datasets/english1970_syntc.csv', header= 0, encoding="utf-8")
df_pos = pd.read_csv('../datasets/english1970_pos.csv', header= 0, encoding="utf-8")
df_punc = pd.read_csv('../datasets/mly1970_punc.csv', header= 0, encoding="utf-8")
df_intj = pd.read_csv('../datasets/english1970_intj.csv', header= 0, encoding="utf-8")
df_idsync = pd.read_csv('../datasets/english1970_idsync.csv', header= 0, encoding="utf-8")
```

```python
# all column label
df_syn.columns = ['sync','label']
df_pos.columns = ['pos','label']
df_punc.columns = ['punc','label']
df_intj.columns = ['intj','label']
df_idsync.columns = ['idsync','sarc_majority']
```

```python
# concatenate all column
df_combine = pd.concat([df_syn, df_pos, df_punc, df_intj, df_idsync], axis=1)
df_combine.head(10)
```

```python
# drop repeatable same column label
df = df_combine.drop('label', 1)
```

```python
# save to by different column
df.to_csv('../datasets/sarc_combine_diff.csv', index=False)
```
